```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
from sklearn.datasets import load_iris
```

```python
iris = pd.read_csv(r"C:\Users\KAsab\Downloads\Iris.csv")
```

```python
iris.head()
```

```python
X = iris.iloc[:,1:5].values
```

```python
y = iris.iloc[:,-1].values.ravel()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.2,random_state = 1)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import make_pipeline
```

```python
scaler = StandardScaler()
```

```python
lr = LogisticRegression()
```

```python
model = make_pipeline(scaler,lr)
```

```python
model.fit(X_train,y_train)
```

```python
lr.fit(X_train,y_train)
```

```python
lr.score(X_test,y_test)
```

```python
X2_train,X2_test,y2_train,y2_test = train_test_split(X,y,test_size = 0.2,random_state = 12)
```

```python
lr.fit(X2_train,y2_train)
```

```python
lr.score(X2_test,y2_test)
```

```python
from sklearn.model_selection import cross_val_score
```

```python
cvs = cross_val_score(model,X,y,cv = 10)
```

```python
cvs
```

```python
np.average(cvs)
```

```python
np.std(cvs)
```

```python
from sklearn.model_selection import KFold
```

```python
kf = KFold(n_splits = 15,shuffle = True,random_state = 123)
```

```python
kf_score = cross_val_score(model, X, y, cv=kf, scoring="accuracy")
```

```python
kf_score
```

```python
np.average(kf_score)
```

```python
np.std(kf_score)
```

```python
from sklearn.model_selection import StratifiedKFold
```

```python
kf2 = StratifiedKFold(n_splits = 10,shuffle = True,random_state = 123)
```

```python
kfscore2 = cross_val_score(model,X,y,cv = kf2)
```
