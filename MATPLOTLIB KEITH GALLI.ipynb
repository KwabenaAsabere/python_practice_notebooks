```python
x = np.linspace(0,10,100)
y1,y2,y3 = np.cos(x),np.cos(x+1),np.cos(x +2)
names = ["Signal 1", "Signal 2","Signal 3"]
```

```python
fig,axes = plt.subplots(nrows =3,figsize = (10,10))
for ax,y,name in zip(axes,[y1,y2,y3],names):
    ax.plot(x,y)
    ax.set(xticks=[],yticks=[],title = name)
plt.show()
```

```python
plt.figure(figsize = (8,6))
plt.title("Gas Prices over Time (in USD)")
plt.plot(gas["Year"],gas["USA"],"r.-",label = "United States")
plt.plot(gas["Year"],gas["Canada"],"b.-",label = "Canada")
plt.plot(gas["Year"],gas["South Korea"],"g.-",label = "South Korea")
plt.plot(gas["Year"],gas["Australia"])
plt.xticks(gas["Year"][::3])
plt.legend()
plt.show()
```

```python
plt.figure(figsize = (8,6))
countries = ["USA","Canada","Australia","South Korea"]
for country in countries:
    plt.plot(gas["Year"],gas[country],marker = "o",label = country)

plt.title("Gas Prices over Time (in USD)", fontdict ={"fontweight":"bold","fontsize":22})
plt.xticks(gas["Year"][::3])
plt.legend()
plt.savefig('Gas_prices.png',dpi= 300)
plt.show()
```

```python
left = fifa.loc[fifa["Preferred Foot"] == "Left"].count().iloc[0]
right = fifa.loc[fifa["Preferred Foot"]== "Right"].count().iloc[0]
labels = ["Left","Right"]
plt.pie([left,right],labels = labels,autopct = "%.2f%%")
plt.title("Foot Preference of FIFA Players")
plt.show()
```

```python
fifa["Weight"] =pd.to_numeric(fifa["Weight"].str.replace("lbs",""),errors = "coerce")
```

```python
#fifa["Weight"] =pd.to_numeric(fifa["Weight"].str.replace("lbs",""),errors = "coerce")
light =fifa.loc[fifa["Weight"]< 125].count().iloc[0]
light_medium = fifa.loc[(fifa["Weight"]>=125) & (fifa["Weight"]<150)].count().iloc[0]
medium =fifa.loc[(fifa["Weight"]>=150) & (fifa["Weight"] < 175)].count().iloc[0]
medium_heavy =fifa.loc[(fifa["Weight"]>= 175) & (fifa["Weight"] < 200)].count().iloc[0]
heavy = fifa.loc[(fifa["Weight"] >= 200)].count().iloc[0]

weights = [light,light_medium,medium,medium_heavy,heavy]
labels = ["Under 125","125-150","150-175","175-200","Over 200"]
explode = [0.4,0.1,0.1,0.1,0.4]
plt.pie(weights,labels = labels,autopct = "%.2f %%", pctdistance = 0.8,explode =explode )
plt.show()
```

```python
plt.figure(figsize = (5,8))
barcelona = fifa.loc[fifa["Club"]== "FC Barcelona"]["Overall"]
madrid =fifa.loc[fifa["Club"]== "Real Madrid"]["Overall"]
revs = fifa.loc[fifa["Club"] == "New England Revolution"]["Overall"]
labels = ["Barcelona","Real Madrid","New England Revolution"]
boxes = plt.boxplot([barcelona,madrid,revs],labels = labels,patch_artist = True,medianprops = {"linewidth":2})
for box in boxes["boxes"]:
    box.set(color ="blue",linewidth = 2 )
    box.set(facecolor  = "lightgrey")
plt.title("Professional Soccer Team Comparison")
plt.ylabel("Overall FIFA Rating")
plt.show()
```

```python
data = np.random.standard_normal(30).cumsum()
data
```

```python
fig,ax = plt.subplots(figsize = (8,6))
ax.plot(data,"k-o",label = "Default")
ax.plot(data,"r--o",drawstyle = "steps-post",label = "Steps-post")
ax.legend()
plt.grid()
plt.show()
```

```python
np.random.seed(123)
fig,ax = plt.subplots(figsize = (8,6))
ax.plot(np.random.standard_normal(1000).cumsum());
ax.axhline(-20,color = "red",linestyle = "dashed",linewidth = 3)
ticks = ax.set_xticks([0,250,500,750,1000])
labels = ax.set_xticklabels(["One","Two","Three","Four","Five"],rotation = 30,fontsize = 8)
ax.set(title = "Standard Plot", xlabel= "Stages")
ax.grid()
```
