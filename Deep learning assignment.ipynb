```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_moons
import pandas as pd

# Generate a basic non-linear dataset
X, y = make_moons(n_samples=300, noise=0.2, random_state=42)

# Simulating additional features related to cardiovascular health
# Age: Random values between 20 and 80
age = np.random.randint(20, 80, size=(X.shape[0], 1))

# Cholesterol: Random values between 150 and 300
cholesterol = np.random.randint(150, 300, size=(X.shape[0], 1))

# Blood pressure: Random values between 90 and 180
blood_pressure = np.random.randint(90, 180, size=(X.shape[0], 1))

# Concatenate additional features to the existing moon dataset
X_health = np.hstack([X, age, cholesterol, blood_pressure])

# Convert to pandas dataframe for easy handling
df = pd.DataFrame(X_health, columns=['Feature1', 'Feature2', 'Age', 'Cholesterol', 'BloodPressure'])
df['Label'] = y

# Display the first few rows
print(df.head())

# Plot the dataset (only the original 2D moon features)
plt.scatter(X[:, 0], X[:, 1], c=y, cmap='coolwarm')
plt.title('Non-linearly Separable Cardiovascular Dataset')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.show()
```
