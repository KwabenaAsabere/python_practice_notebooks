```python
weather = pd.read_csv(r"https://raw.githubusercontent.com/chendaniely/scipy-2020-pandas/master/data/weather.csv")
```

```python
 ebola_long = ebola.melt(id_vars=["Date","Day"],var_name="cd_country",value_name = "count")
ebola_long.head()
```

```python
ebola_long[["satus","country"]]= ebola_long["cd_country"].str.split("_",expand= True)
ebola_long.head()
```

```python
weather.head()
```

```python
weather_long = weather.melt(id_vars= ["id","year","month","element"],var_name = "day",value_name="temp")
weather_long.head()
```

```python
weather_long.dropna(inplace= True)
```

```python
weather_long.head()
```

```python
weather_tidy =(weather_long
               .pivot_table(index = ["id","year","month","day"],
                            columns="element",
                            values ="temp")
               .reset_index())
```

```python
weather_tidy.head(10)
```

```python
weather_tidy.sort_values(["year","month","day"]).head()
```

```python
weather_tidy["day"] = weather_tidy["day"].str.replace("d","")
```

```python
weather_tidy.head()
```

```python
weather_tidy.to_csv("weather_clean.csv",index=False,sep = ",")
```
