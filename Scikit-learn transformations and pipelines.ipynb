### Advanced Pipeline

```python
penguins.head()
```

```python
X = penguins.iloc[:,0:6]
y = penguins.iloc[:,6]
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.3,random_state = 123)
```

```python
num_cols = ["bill_length_mm","bill_depth_mm","flipper_length_mm","body_mass_g"]
cat_cols = ["island","species"]
```

```python
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
```

```python
num_pipeline = Pipeline(
   steps= [
    ("impute",SimpleImputer(strategy = "mean")),
    ("scale",StandardScaler())
   ] )
```

```python
cat_pipeline = Pipeline(steps = [
    ("impute",SimpleImputer(strategy = "most_frequent")),
    ("one_hot_encoder",OneHotEncoder(handle_unknown = "ignore",sparse_output = False))
])
```

```python
col_transformer = ColumnTransformer(transformers = [
    ("num_pipeline",num_pipeline,num_cols),
    ("cat_pipeline",cat_pipeline,cat_cols)
],
  remainder = "drop",
   n_jobs =-1                                )
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dtc = DecisionTreeClassifier()
```

```python
pipefinal = make_pipeline(col_transformer,dtc)
```

```python
pipefinal.fit(X_train,y_train)
```

```python
pipefinal.score(X_test,y_test)
```

```python
housing = pd.read_csv(r"C:\Users\KAsab\Downloads\Compressed\archive_37\california_housing.csv")
```

```python
X = housing.iloc[:,:-2].to_numpy()
X
```

```python
y= housing.iloc[:,-2].to_numpy()
```

```python
y
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size = 0.3,random_state = 123)
```

```python
X_train.shape,X_test.shape,y_train.shape,y_test.shape
```

```python
from sklearn.preprocessing import MinMaxScaler,StandardScaler,FunctionTransformer
from copy import deepcopy
```

```python
std_scaler = StandardScaler().fit(X_train[:,:2])
mms = MinMaxScaler().fit(X_train[:,2:])
```

```python
def preprocessor(X):
    A = np.copy(X)
    A[:,:2]= std_scaler.transform(X[:,:2])
    A[:,2:]= mms.transform(X[:,2:])
    return A
```

```python
preprocessor(X_test)
```

```python
preprocess_transformer = FunctionTransformer(preprocessor)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
```

```python
p1 = Pipeline(steps = [
    ("Impute",SimpleImputer(strategy = "mean")),
    ("Scaler",preprocess_transformer),
("Linear Regression",lr),
])
```

```python
p1
```

```python
from sklearn.metrics import mean_absolute_error
```

```python
def fit_and_print(p,X_train = X_train,y_train = y_train,X_test= X_test,y_test= y_test):
    p.fit(X_train,y_train)
    train_preds = p.predict(X_train)
    test_preds = p.predict(X_test)
    print("Training error: " + str(mean_absolute_error(train_preds,y_train)))
    print("Test error: " + str(mean_absolute_error(test_preds,y_test)))
```

```python
fit_and_print(p1)
```

```python
from sklearn.neighbors import KNeighborsRegressor  as KNR
```

```python
p2 = Pipeline(steps=[
    ("Imputer",SimpleImputer(strategy = "mean")),
    ("Scaler",preprocess_transformer),
    ("KNN Regression",KNR(n_neighbors = 7))
])
fit_and_print(p2)
```
