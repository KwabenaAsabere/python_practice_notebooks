```python
inertia = []
K = range(1,11)
```

```python
for k in K:
    kmeans = KMeans(n_clusters = k,random_state = 42)
    kmeans.fit(features_std)
    inertia.append(kmeans.inertia_)
```

```python
plt.figure(figsize = (8,6))
plt.plot(K,inertia,marker = "o")
plt.xlabel("Number of Clusters (k)")
plt.ylabel("Inertia")
plt.title("Elbow Method for Optimal k")
plt.show()
```

```python
K = range(2,11)
```

```python
K = range(2,11)
silhouette_scores = []
for k in K:
    kmeans = KMeans(n_clusters = k,random_state = 42)
    labels = kmeans.fit_predict(features_std)
    score = silhouette_score(features_std,labels)
    silhouette_scores.append(score)
```

```python
plt.figure(figsize = (8,6))
plt.plot(K,silhouette_scores,marker = "o")
plt.xlabel("Number of clusters (k)")
plt.ylabel("Silhouette Score")
plt.title("Silhouette Score for Optimal k")
plt.show()
```

```python
kmeans = KMeans(n_clusters = 3,random_state = 42)
kmeans.fit(features_std)
y_kmeans = kmeans.predict(features_std)
centers = kmeans.cluster_centers_
print("Cluster Centers:\n", centers)
```

```python
pca = PCA(n_components = 2)
features_pca = pca.fit_transform(features_std)
plt.figure(figsize = (8,6))
plt.scatter(features_pca[:,0],features_pca[:,1], c= y_kmeans,cmap = "viridis",
           label = "Clusters")
plt.scatter(pca.transform(centers)[:,0],pca.transform(centers)[:,1], c= "red", s= 200,label = "Centroids",
           marker = "X")
plt.xlabel("Principal Component 1")
plt.ylabel("Principal Component 2")
plt.title("Kmeans Clustering on Wisconsin Dataset (PCA-reduced)")
plt.legend()
plt.show()
```

```python
score = silhouette_score(features_std,y_kmeans)
print("Silhouette Score:",score)
```
