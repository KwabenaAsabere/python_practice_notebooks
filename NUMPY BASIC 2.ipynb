```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
rng = np.random.default_rng(seed = 12345)
```

```python
a=rng.standard_normal(10000)
```

```python
plt.hist(a,edgecolor = "black",bins = 50)
plt.grid()
plt.show()
```

```python
np.random.randint(100,size = (20))
```

```python
x = rng.normal(loc = 10,scale =2,size = 100000)
```

```python
plt.hist(x, edgecolor = "black",color = "grey",bins = 30,density = True)
plt.grid()
plt.xlabel("Values")
plt.ylabel("Frequency")
plt.show()
```

```python
x_binom=rng.binomial(100,p = 0.5,size = 100000)
```

```python
plt.hist(x_binom,bins = 50,edgecolor = "black",color = 'grey');
```
