```python
from sklearn.pipeline import make_pipeline
```

```python
from sklearn.compose import make_column_transformer
```

```python
from sklearn.preprocessing import StandardScaler,OneHotEncoder,MinMaxScaler
```

```python
from sklearn.model_selection import train_test_split
```

```python
ohe = OneHotEncoder(sparse_output = False)
ss = StandardScaler()
```

```python
toyota.head(0)
```

```python
X = toyota.iloc[:,1:]
y = toyota.iloc[:,0]
```

```python
cat_cols = ["FuelType"]
num_cols = X.select_dtypes(include = ["number"]).columns.tolist()
```

```python
print(num_cols)
```

```python
print(cat_cols)
```

```python
ct = make_column_transformer(
    (ohe,cat_cols),
    (ss,num_cols),
    remainder = "drop"
)
```

```python
ct.set_output(transform = "pandas")
```

```python
X = ct.fit_transform(X)
```

```python
X.head()
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y)
```

```python
print(X_train.shape)
print(y_train.shape)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
```

```python
from sklearn.impute import SimpleImputer
```

```python
imputer = SimpleImputer()
```

```python
pipe = make_pipeline(imputer,lr)
```

```python
pipe
```

```python
pipe.fit(X_train,y_train)
```

```python
pipe.score(X_train,y_train)
```

```python
pipe.score(X_test,y_test)
```

```python
lr.coef_
```

```python
X_cols = X.columns.str.split("__").str.get(1).tolist()
print(X_cols)
```

```python
cdf = pd.DataFrame(lr.coef_,X_cols,columns = ["Coeficient"])
print(cdf)
```

```python
y_preds = lr.predict(X_test)
```

```python
residuals = y_test - y_preds
```

```python
fig,ax = plt.subplots()
ax.scatter(y_preds,y_test)
plt.xlabel("Predicted Values")
```

```python
from sklearn.metrics import mean_absolute_error,mean_squared_error, r2_score
```

```python
print("Mean Absolute Error :",mean_absolute_error(y_test,y_preds))
print("Mean Squared Error :",mean_squared_error(y_test,y_preds))
print("R-Squared :",r2_score(y_test,y_preds))
print("RMSE :",np.sqrt(mean_squared_error(y_test,y_preds)))
```

```python
sns.displot(residuals,kde = True)
```

```python
sns.scatterplot(x =y_preds,y=residuals)
plt.axhline(0,color = "red",linestyle = "dashed")
plt.xlabel("Predicted")
plt.ylabel("Residuals")
```

```python
sns.residplot(x = y_preds, y = residuals,lowess = True,line_kws={"color":"red"})
```
