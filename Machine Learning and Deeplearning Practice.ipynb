```python
Default = pd.read_csv(r"C:\Users\KAsab\Desktop\PYTHON\ISLP DATASETS\Default.csv")
```

```python
Default.shape
```

```python
sns.boxplot(
    data = Default,
    x = "default",
    y = "balance",
    hue = "default"
)
plt.xlabel("Default")
plt.ylabel("Balance")
```

```python

fig,ax  = plt.subplots(figsize = (6,6))
sns.boxplot(
    data = Default,
    x = "default",
    y = "income",
    hue = "default",
    ax = ax
)
ax.set_xlabel("Default")
ax.set_ylabel("Income")
```
