```python
num_cols = cva.iloc[:,[0,5]].columns.tolist()
cat_cols = cva.iloc[:,[1,2,3,4]].columns.tolist()
```

```python
print(num_cols)
print(cat_cols)
```

```python
from sklearn.pipeline import Pipeline
```

```python
num_pipeline = Pipeline(steps = [
                        ("impute",SimpleImputer(strategy = "mean")),
                        ("scale",StandardScaler())
])
```

```python
cat_pipeline = Pipeline(steps = [
    ("impute",SimpleImputer(strategy = "most_frequent")),
    ("one_not-encoder",OneHotEncoder(handle_unknown = "ignore",sparse_output = False))
])
```

```python
from sklearn.compose import ColumnTransformer
```

```python
col_transformer = ColumnTransformer(transformers=[
    ("num_pipeline",num_pipeline,num_cols),
    ("cat_pipeline",cat_pipeline,cat_cols)
],
                                    remainder = "drop",
                                    n_jobs = -1
                                   )
```

```python
col_transformer
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dtc = DecisionTreeClassifier()
```

```python
X = cva.iloc[:,:6]
y = cva.iloc[:,-1]
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y)
```

```python
X_train
```

```python
pipefinal = make_pipeline(col_transformer,dtc)
```

```python
pipefinal.fit(X_train,y_train)
```

```python
pipefinal.score(X_test,y_test)
```
