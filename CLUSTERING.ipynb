```python

from sklearn.cluster import AgglomerativeClustering
import scipy.cluster.hierarchy as sch
```

```python
# Perform hierarchical clustering using scikit-learn's AgglomerativeClustering
clustering = AgglomerativeClustering(n_clusters=3, metric='euclidean', linkage='ward')
clustering.fit(genes)
```

```python

# View the cluster labels assigned to each sample
print(clustering.labels_)
```

```python
# Optional: Plot the dendrogram for visualization
linked = sch.linkage(genes, method='ward')
plt.figure(figsize=(10, 7))
sch.dendrogram(linked)
plt.title('Hierarchical Clustering Dendrogram')
plt.xlabel('Sample index')
plt.ylabel('Distance')
plt.show()
```
